# Playing with Windows

```python
import numpy as np
from scipy import signal
import scipy
from matplotlib import pyplot as plt
```

```python
n = 201
rect = signal.windows.boxcar(n)
flat = signal.windows.flattop(n)
bart = signal.windows.bartlett(n)
```

```python
npts = 2**13

Rect = np.abs(scipy.fft.fft(np.hstack((rect, np.zeros(npts-n)))))
Flat = np.abs(scipy.fft.fft(np.hstack((flat, np.zeros(npts-n)))))
Bart = np.abs(scipy.fft.fft(np.hstack((bart, np.zeros(npts-n)))))

f = np.linspace(0,1, npts)
T = 1/f
```

```python
plt.plot(Rect)
```

```python
plt.plot(T,Rect)
plt.plot(T, Flat)
plt.plot(T, Bart)
#plt.yscale('log')
#plt.xscale('log')

plt.xlim([20,30])
plt.grid()
```
