# Look at depth data

```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
```

```python
fn = '/datadrive/ooi-rs03ccal-mj03f-05-botpta301_fce5_0939_5f02.nc'
ds = xr.open_dataset(fn)

fn2 = '/datadrive/ooi-rs03ecal-mj03e-06-botpta302_3cd9_ef7a_abf3.nc'
ds2 = xr.open_dataset(fn2)
```

```python
cc = xr.DataArray(ds['botpres'], dims=['time'], coords={'time':ds['time'].values}, name='bottom_pressure')
ec = xr.DataArray(ds2['botpres'], dims=['time'], coords={'time':ds2['time'].values}, name='bottom_pressure')
```

```python
plt.figure(figsize=(13,3))
(0.689476*cc).rolling(time=1001*60, min_periods=100).mean().plot()
(0.689476*ec).rolling(time=1001*60, min_periods=100).mean().plot()
plt.grid()
```

```python
plt.figure(figsize=(13,3))
(0.689476*(ec-cc)).rolling(time=1001*60, min_periods=100).mean().plot()
plt.grid()
```
